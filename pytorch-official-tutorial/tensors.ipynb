# Official PyTorch tutorial

## Tensors

This Jupyter notebook is self explanatory. No need to document or add more information.

```python
import torch
import numpy as np
```

```python
# Create tensors directly from data
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print("Tensor from data:\n", x_data)
```

```python
# From a numpy array
np_array = np.array(data)
x_array = torch.from_numpy(np_array)
print("Tensor from numpy array:\n", x_array)
```

```python
# From another tensor
x_ones = torch.ones_like(x_data)
print("Ones Tensor:\n", x_ones)

x_rand = torch.rand_like(x_data, dtype=torch.float)
print("Random tensor:\n", x_rand)
```

## Shape

```python
shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random tensor: \n{rand_tensor}\n")
print(f"Ones tensor: \n{ones_tensor}\n")
print(f"Zeros tensor: \n{zeros_tensor}\n")
```

## Tensor attributes

```python
tensor = torch.rand(3, 4)

print(f"Shape of tensor: \n{tensor.shape}\n")
print(f"Datatype of tensor: \n{rand_tensor.dtype}\n")
print(f"Device tensor is stored on: \n{rand_tensor.device}\n")
```

## Tensor Operations

```python
# We move our tensor to the GPU if available
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
    print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Standard numpy-like indexing and slicing
tensor = torch.ones(4, 4)
tensor[:,1] = 0
print(tensor)
```

```python
# Concatenate tensors
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
## Multiplying tensors
# Compute element-wise product
print(f"tensor.mul(tensor)\n {tensor.mul(tensor)}\n")
# Alternative syntax
print(f"tensor * tensor \n {tensor * tensor}")
```

```python
## Matrix multiplication

print(f"tensor.matmul(tensor.T) \n {tensor.matmul(tensor.T)}\n")
print(f"tensor @ tensor.T \n {tensor @ tensor.T}")
```

```python
## In place operations: operations that have a _ suffix are in place.
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

## Bridge with numpy

Tensors on the CPU and NumPy arrays can share their underlying memory locations, and changing one will change the other

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
# A change in the tensor reflects in the NumPy array
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
# Numpy arrays to tensors
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
